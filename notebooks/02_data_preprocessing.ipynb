# 02 Data Pre-Processing

```python
import os
from pathlib import Path
import pandas as pd
parent_path=Path.cwd().parent
```

```python
train_path=parent_path/"data/raw/train.csv"
```

```python
df=pd.read_csv(train_path)
df.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
import joblib
def preprocessing(dataframe):
    num_cols=[]
    for col in dataframe.columns:
        if((dataframe[col].dtype=='int64' or dataframe[col].dtype=='float64' ) and (col!='Calories' and col!='id')):
            num_cols.append(col)
            
    #Standardizing the numerical values
    scaler=StandardScaler()
    dataframe[num_cols]=scaler.fit_transform(dataframe[num_cols])
    
    #One-Hot Encoding
    sex_encoded=pd.get_dummies(dataframe['Sex'],prefix='',prefix_sep='')
    
    #Add the encoded columns
    df_encode=pd.concat([dataframe,sex_encoded],axis=1)

    #Let's save this scaler for future reference
    joblib.dump(scaler,filename=Path.cwd().parent/"models/01_scaler.joblib")
    return df_encode
    
            
```

```python
preprocessing(df)
```
